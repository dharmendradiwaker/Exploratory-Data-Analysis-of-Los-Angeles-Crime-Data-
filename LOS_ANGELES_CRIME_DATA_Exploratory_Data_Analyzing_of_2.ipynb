EDA is a philosophy that allows data analysts to approach a database without assumptions. When a data analyst employs EDA, it's like they're asking the data to tell them what they don't know.

## **Q6. Which age group is the largest number of victims?**

To answer this question , we'll use `Vict Age` columns and we will create a new column of age group

```python
# create a list of ages
ages= data_df['Vict Age']

# define the age group ranges
age_bins = [0,20,40,60,80,100]
age_labels=['<20','21-40','41-60','61-80','80+']

#create a dataframe with age column
age_group = pd.DataFrame({'age':ages})

# create the age group column
data_df['age_group']=pd.cut(age_group['age'], bins=age_bins,labels=age_labels)

# display the data
print(data_df['age_group']);
```

```python
victim_age= data_df['age_group'].value_counts(normalize=True)
victim_age
```

```python
victim_age_group = victim_age.reset_index().rename({'index':'age group','age_group':'percentage'},axis=1)
victim_age_group
```

From the above data, we can say that the more than **50% victims are under the age group of 21-40.So, let's visualize it.**

```python
fig=px.pie(
    victim_age_group,
    names='age group',
    values='percentage',
    color_discrete_sequence=px.colors.sequential.Plasma,
    hole=.4,
    title='Victim age group')
fig.show()
```

The above visulisation refers that the most of the victim belongs to the 21-40 age group.

### **Q10 show the number of cases of every victim according to their gender and age group?**

To answer this question , we need to calculate total number victim according to their gender and age group

```python
# count the victim using groupby
vict_gender_age= data_df.groupby(['Vict Sex'])['age_group'].value_counts()
vict_gender_age
```

Here we got the number of victims according to their age group and gender. But this data look litle messy to understand. Let's write a code to generate a dataframe from the above series.

```python
# creating a dictionary of lists from 'vict_gender_age' series
age=[]
sex=[]

for x in range(0, len(vict_gender_age)):
  age.append(vict_gender_age.index[x][0])
  sex.append(vict_gender_age.index[x][1])
age_gender= {
    'gender':age,
    'age group':sex,
    'counts': vict_gender_age
}
```

```python
# generating a dataframe from age_gender dictionary
gender_age_group=pd.DataFrame(age_gender)
gender_age_group
```

Let's form a pivot table using `.pivot` method

```python
# create a pivot table from the gender_age_group dataframe
gender_age_pivot=gender_age_group.pivot('gender','age group','counts');
gender_age_pivot
```

In the above data we can see that the female victims from the age group 21-40 are higher than the males victims.on the other hand, male victims from age group 41-60 are higher in numbers as compare to female victims.
